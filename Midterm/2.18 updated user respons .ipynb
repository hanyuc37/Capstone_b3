# Top Invovled Users Analysis

```python
# df_count shows the total number of resposn per User ID 
df_count = df.groupby('User Id').size().reset_index(name='counts').sort_values(by=['counts'],ascending=False,).reset_index()
df_count.drop(columns="index", inplace = True)
df_count
```

```python
# The top 20 users who respons the most
df_count.head(20)
```

```python
df_count['User Id'].head(20)
```

```python
# The disttubution of total respons per user 
n, bins, patches = plt.hist(x=df_count['counts'], bins= 3, color='#FEA003',
                            alpha=0.7, rwidth=0.85)
plt.grid(axis='y', alpha=0.75)
plt.xlabel('Total respons per users')
plt.ylabel('Counts of Total respons')
plt.title('Counts of Total Respons per user')
maxfreq = n.max()
# Set a clean upper y-axis limit.
plt.ylim(ymax=np.ceil(maxfreq / 10) * 10 if maxfreq % 10 else maxfreq + 10)
```
