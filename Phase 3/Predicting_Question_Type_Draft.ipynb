```python
# setup the BERT Model 
model = Classifier(base_model=BERT, low_memory_mode=True, n_epochs=3, batch_size=100)
```

```python
# setup the data
X = df.Quesh.to_list()
y = df['Avg Rating']


# finetune the model
model.fit(X, y)
```
