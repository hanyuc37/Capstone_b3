```python
import pandas as pd
import seaborn as sns
import numpy as np
import spacy
from spacy import cli
from spacy import displacy
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/drive/MyDrive/Capstone B3/Phase 3/datasets/sentiment_updated.csv") 
```

```python
df.head()
```

```python
df['sent_ans'].describe()
```

```python
# threshold: the mean of the sentiment score as neutral questions --> 0-1.87
# Negative: below 0
# Positive: 1.87+
sent_cat=[]
for i in df['sent_quesh']:
  if i<0:
    sent_cat.append("Negative")
  elif i>=0 and i<=1.87:
    sent_cat.append("Neutral")
  else:
    sent_cat.append("Positive")
```

```python
df['sent_cate']=sent_cat
```

```python
#Distribution of the sentiment quesh category
df.groupby('sent_cate')['sent_cate'].count()
# Most of the questions are Neutral :)
```
