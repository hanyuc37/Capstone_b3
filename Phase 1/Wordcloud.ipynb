```python
from wordcloud import WordCloud
```

```python
def wordcloud(text):
    wc = WordCloud(background_color="white")
    st = ''
    for i in text:
        st+= i
    wordcloud = wc.generate(st)
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.show()
```
